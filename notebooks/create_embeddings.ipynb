```python
import os
import pandas as pd
data_path = '../data'
filename = "text_fields.csv"
fpath_input = os.path.join(data_path,filename)
df = pd.read_csv(fpath_input)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
model.encode(['hello'])
```

```python
from tqdm import tqdm

for i in tqdm(df.itertuples()):
    print([f'{i.text_field}'])
```

```python
# lets create embeddings for these chats
from time import sleep
from tqdm import tqdm
from collections import defaultdict
embeddings = []
payload = []
for i in tqdm(df.itertuples()):
    sentence = [f"{i.text_field}"]
    res = model.encode(sentence)
    embeddings.append(res)
    payload.append({"chat":f"{i.text_field}"})
```

```python
i = 1
y_test = model.encode(payload[i]['chat'])
y_pred = embeddings[i]
y_test = y_test.reshape(-1).tolist()
y_pred = y_pred.reshape(-1).tolist()
```
